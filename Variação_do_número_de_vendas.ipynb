```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
lojas = ['A', 'B', 'C', 'D']

vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}
```

```python
vendas_2022 = pd.DataFrame(vendas_2022, index=lojas)
```

```python
display(vendas_2022)
```

```python
fig, axs = plt.subplots(2,2, figsize=(14,9))
fig.suptitle("Vendas em todo o período de 2022")
plt.subplots_adjust(wspace=0.3, hspace=0.4)

axs[0,0].plot(vendas_2022.loc['A'])

axs[0,1].plot(vendas_2022.loc['B'])

axs[1,0].plot(vendas_2022.loc['C'])

axs[1,1].plot(vendas_2022.loc['D'])

cores=['darkviolet', 'violet', 'green', 'darkblue']
for i, ax in enumerate(axs.flat):
  ax.plot(vendas_2022.loc[vendas_2022.index[i]], color=cores[i], lw=3)
  ax.set_title(f'Vendas da empresa {vendas_2022.index[i]}', loc='left', fontsize=13)
  ax.set_xlabel('Mês', fontsize=14)
  ax.set_ylabel('Número de vendas', fontsize=14)

  ax.tick_params(labelsize=9)
  ax.grid(color='lightgrey')

plt.show()
```
